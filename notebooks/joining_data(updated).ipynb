Join the data from Part 1 with the data from Part 2 to create a new dataframe.

```python
import pandas as pd

# Read the two Excel files into two separate DataFrames
df1 = pd.read_excel('hamilton_bike_stations_most_spots.xlsx')
df2 = pd.read_excel('foursquare_multi_coordinates_data_with_original.xlsx')

# Merge DataFrames based on latitude and longitude columns
merged_df = pd.merge(df1, df2, left_on=['original_latitude', 'original_longitude'], right_on=['original_latitude', 'original_longitude'], how='inner')

# Now merged_df contains the data from both Excel files in a single DataFrame, merged based on latitude and longitude
```

```python
print(merged_df)
```

```python
# Save the merged DataFrame to a new CSV file
merged_df.to_csv('merged_data.csv', index=False)
```

Provide a visualization that you used as part of your EDA process. Explain the initial pattern or relationship you discoved through this visualization. 

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Assume 'merged_df' is your DataFrame containing the merged data

# Filter data to get the bike station with the most empty slots
most_empty_station = merged_df.loc[merged_df['empty_slots'].idxmax()]

# Bar plot comparing empty slots for each station with business names
plt.figure(figsize=(12, 8))
barplot = sns.barplot(x='station_name', y='empty_slots', data=merged_df, ci=None)
plt.bar(most_empty_station['name'], most_empty_station['empty_slots'], color='red', label='Most Empty Slots')

# Add labels and title
plt.xlabel('Bike Station or Business Name')
plt.ylabel('Number of Empty Slots')
plt.title('Comparison of Empty Slots for Each Station with Nearby Business Names')

# Rotate x-axis labels for better readability
plt.xticks(rotation=45, ha='right')

# Add a legend
plt.legend()

# Show the plot
plt.show()
```

```python
import pandas as pd

# Read the three Excel files into three separate DataFrames
df1 = pd.read_excel('hamilton_bike_stations_most_spots.xlsx')
df2 = pd.read_excel('foursquare_multi_coordinates_data_with_original.xlsx')
df3 = pd.read_excel('yelp_businesses_multiple_locations.xlsx')

# Merge the first two DataFrames based on latitude and longitude columns
merged_df = pd.merge(df1, df2, left_on=['original_latitude', 'original_longitude'], right_on=['original_latitude', 'original_longitude'], how='inner')

# Merge the resulting DataFrame with the third DataFrame based on latitude and longitude columns
merged_df = pd.merge(merged_df, df3, left_on=['original_latitude', 'original_longitude'], right_on=['original_latitude', 'original_longitude'], how='inner')

# Now merged_df contains the data from all three Excel files in a single DataFrame, merged based on latitude and longitude
```

```python
print(merged_df)
```

```python
# Save the merged DataFrame to a new CSV file
merged_df.to_csv('merged_data.csv', index=False)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Assume 'merged_df' is your DataFrame containing the merged data

# Set up the figure and axes
plt.figure(figsize=(12, 8))

# Scatter plot with linear regression for each business
lmplot = sns.lmplot(x='rating', y='empty_slots', hue='name_y', data=merged_df, height=8, aspect=1.5, ci=None)

# Add labels and title
plt.xlabel('Your Predictor Variable')
plt.ylabel('Number of Empty Slots')
plt.title('Linear Regression between Predictor Variable and Empty Slots')

# Show the plot
plt.show()
```
