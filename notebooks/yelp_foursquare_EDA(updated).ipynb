# Yelp

Send a request to Yelp with a small radius (1000m) for all the bike stations in your city of choice. 

```python
url = "https://api.yelp.com/v3/businesses/search"

headers = {
    "accept": "application/json",
    "Authorization": #########################
}

params = {
    "term": "Bike Station Bike Sharing",  
    "location": "Hamilton",  
    "radius": 1000,  # Radius in meters
    "sort_by": "best_match",
    "limit": 20  # Adjust as needed
}

response = requests.get(url, headers=headers, params=params)

if response.status_code == 200:
    data = response.json()
    businesses = data.get('businesses', [])

    for business in businesses:
        print("Name:", business.get('name'))
        print("Rating:", business.get('rating'))
        print("Address:", ', '.join(business.get('location', {}).get('display_address', [])))
        print("Categories:", ', '.join([category['title'] for category in business.get('categories', [])]))
        print("\n")

else:
    print(f"Error: {response.status_code}")
    print(response.text)
```
