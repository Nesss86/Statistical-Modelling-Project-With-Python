# CityBikes

Send a request to CityBikes for the city of your choice. 

```python
import requests

# Specify the base URL for the CityBikes API
base_url = "http://api.citybik.es/v2/networks"

# Send a GET request to the base URL
response = requests.get(base_url)

# Display all headers
print("Response Headers:")
for header, value in response.headers.items():
    print(f"{header}: {value}")

# Handling the response
if response.status_code == 200:
    # Process the response as needed
    print("\nResponse Content:")
    print(response.json())
else:
    print(f"\nError: {response.status_code}")
```

```python
import requests

# Specify the URL for listing all networks with specific fields
url = "http://api.citybik.es/v2/networks?fields=id,name,href"

# Send a GET request
response = requests.get(url)

# Handling the response
if response.status_code == 200:
    networks = response.json()["networks"]
    for network in networks:
        print(f"ID: {network['id']}, Name: {network['name']}, API Endpoint: {network['href']}")
else:
    print(f"Error: {response.status_code}")
```
